```python
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df["target"] = iris.target
df
```

```python
sns.scatterplot(data=df, x='petal length (cm)', y='petal width (cm)', hue='target', palette='bright')
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0, max_iter=500)
clf.fit(df.drop(["target"], axis=1), df["target"])
y_pred = clf.predict(df.drop(["target"], axis=1))
```

```python
df["y_pred"] = y_pred
sns.scatterplot(data=df, x="petal length (cm)", y="petal width (cm)", hue='y_pred', palette='bright')
```
