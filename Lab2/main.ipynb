```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("housing.csv", header=None, delimiter=r'\s+')
y = data[13]
```

## 1. Увеличение количества признаков в датасете

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
```

```python
# 2 признака
X = data.iloc[:,0:2]
lr = LinearRegression()
lr.fit(X, y)

y_pred = lr.predict(X)

print(f"Case 2")
print(f"MAE: {mean_absolute_error(y, y_pred)}")
print(f"R2: {r2_score(y, y_pred)}")
print(f"MSE: {mean_squared_error(y, y_pred)}")
```

### Построение модели для пункта 2

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

x1 = X.iloc[:, 0]
x2 = X.iloc[:, 1]

ax.scatter(x1, x2, y)

x1_range = np.linspace(x1.min(), x1.max(), 20)
x2_range = np.linspace(x2.min(), x2.max(), 20)
x1_grid, x2_grid = np.meshgrid(x1_range, x2_range)

x_grid_flat = np.c_[x1_grid.ravel(), x2_grid.ravel()]
y_grid_pred = lr.predict(x_grid_flat).reshape(x1_grid.shape)

ax.plot_surface(x1_grid, x2_grid, y_grid_pred, alpha=0.6, color="green")

ax.set_xlabel('Первый признак')
ax.set_ylabel('Второй признак')

plt.show()
```

### Продолжение

```python
# 7 признаков
X = data.iloc[:,0:7]
lr = LinearRegression()
lr.fit(X, y)

y_pred = lr.predict(X)

print(f"Case 7")
print(f"MAE: {mean_absolute_error(y, y_pred)}")
print(f"R2: {r2_score(y, y_pred)}")
print(f"MSE: {mean_squared_error(y, y_pred)}")
```

```python
# 13 признаков
X = data.iloc[:,0:13]
lr = LinearRegression()
lr.fit(X, y)

y_pred = lr.predict(X)

print(f"Case 13")
print(f"MAE: {mean_absolute_error(y, y_pred)}")
print(f"R2: {r2_score(y, y_pred)}")
print(f"MSE: {mean_squared_error(y, y_pred)}")
```

Получается, что увеличение количества признаков привело к увеличению качества модели

## 2. Увеличение количества данных

```python
from sklearn.model_selection import train_test_split
```

```python
X = data[range(13)]
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.2, random_state=0)
lr = LinearRegression() 
lr.fit(X_train, Y_train)

y_pred = lr.predict(X)

print(f"MAE: {mean_absolute_error(y, y_pred)}")
print(f"R2: {r2_score(y, y_pred)}")
print(f"MSE: {mean_squared_error(y, y_pred)}")
```
